## Understanding Data

### Attribute Information
- Age: age of the patient (in yeara)
- Sex: sex of the patient (M: Male, F: Female)
- ChestPainType: chest pain type (TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic)
- RestingBP: resting blood pressure (in mm Hg)
- Cholesterol: serum cholesterol (in mm/dl)
- FastingBS: fasting blood sugar (1: if FastingBS > 120 mg/dl, 0: otherwise)
- RestingECG: resting electrocardiogram results (Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria)
- MaxHR: maximum heart rate achieved
- ExerciseAngina: exercise-induced angina (Y: Yes, N: No)
- Oldpeak: oldpeak = ST (Numeric value measured in depression)
- ST_Slope: the slope of the peak exercise ST segment (Up: upsloping, Flat: flat, Down: downsloping)
- HeartDisease: output class (1: heart disease, 0: Normal)

## Correlation Matrix

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df.select_dtypes(include=['float64', 'int64']).corr()[['HeartDisease']].sort_values(by='HeartDisease', ascending=False),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation of Features with Target (HeartDisease)")
plt.show()
```

#### Insights
- Oldpeak value and HeartDisease are positively correlated i.e. Higher the ST depression on ECG (in mm) the higher are the chances of CVDs
- MaxHR and HeartDisease are negatively correlated i.e. The lower the HeartRate, the higher the the chances of CVDs
- Low cholesterol may also result in CVDs
- Having Higher FastingBP may result in CVDs

# Univariate Analysis

```python
df.columns
```

```python
def countplot_with_percent(data, column, palette='Set3'):
    plt.figure(figsize=(5, 4))
    ax = sns.countplot(x=column, data=data, hue=column, palette=palette, legend=False)

    total = len(data)
    
    for p in ax.patches:
        count = p.get_height()
        percent = f'{100 * count / total:.1f}%'
        ax.annotate(percent, (p.get_x() + p.get_width() / 2., count),
                    ha='center', va='bottom', fontsize=10, color='black')
    
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

cat_cols = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope', 'HeartDisease']

for col in cat_cols:
    countplot_with_percent(df, col)
```

#### Insights
- The data consists of mostly men (79%)
- Most people suffer with Asymptomatic chest pain (i.e. absence of chest pain)
- Most people have normal FastingBS
- Most people have normalECG
- Almost 60% people did not expiriece chest pain induced due to exercise
- Most people have Up or Flat ST Slope which is not usually concerned with CVDs
- More that half of the people have CVDs (55%)

```python
colors = ['brown', 'blue', 'orange', 'red', 'green']
num_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

for i, col in enumerate(num_cols):
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    
    sns.histplot(df[col], kde=True, color=colors[i], ax=axes[0])
    axes[0].set_title(f'Histogram of {col}')
    
    sns.boxplot(x=df[col], color=colors[i], ax=axes[1])
    axes[1].set_title(f'Boxplot of {col}')
    
    plt.tight_layout()
    plt.show()
```

#### Insights
- Age is almost normally distributed and have no outliers
- RestingBP is negatively skewed with outliers: low values like 0 and high values like 200 indicate counting error
- Cholestrol have make counting errors as it cannot be 0. It is positively skewed
- MaxHR is almost mormally distributed and have few outliers
- Oldpeak deviates a lot from normal distribution. It contains outliers which may be helpful to diagnose CVDs

# Bivariate Analysis

```python
num_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

for col in num_cols:
    plt.figure(figsize=(10, 4))

    # Boxplot
    plt.subplot(1, 2, 1)
    sns.boxplot(x='HeartDisease', y=col, data=df, palette='pastel')
    plt.title(f'Boxplot: {col} vs HeartDisease')

    # Barplot
    plt.subplot(1, 2, 2)
    plt.title(f'Barplot: {col} vs HeartDisease')
    sns.barplot(x='HeartDisease', y=col, data=df, palette='Set2')

    plt.tight_layout()
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(10, 4))
    # Violinplot
    plt.subplot(1, 2, 2)
    sns.violinplot(x='HeartDisease', y=col, data=df, palette='Set3')
    plt.title(f'Violinplot: {col} vs HeartDisease')
```

#### Insights
- Heart Disease is seem more in people high more age
- RestingBP does not show much effect on chance of CVDs
- The data is inconsistent/defective for Cholesterol, as most people with 0 cholesterol (error value) have heart disease
- People with lower MaxHR have higher chances of heart disease
- People with Oldpeak value near 0 are least likely to have heart disease

```python
cat_cols = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, hue='HeartDisease', data=df, palette='bright')
    plt.title(f'{col} vs HeartDisease')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

#### Insights
- Males are more likely to have heart disease
- People with ASY chest pain are most likely to have CVD
- People with higher FastingBS are more likely to have heart disease
- People with ST-T wave abnormality more likely to have heart disease
- People who suffers from exercise angina are more likely to have heart issues
- People with ST slope Flat or Down are more likely to have CVDs
